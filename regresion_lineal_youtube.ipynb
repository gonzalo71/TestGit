### Boston house price prediction
The problem that we are going to solve here is that given a set of features that describe a house in Boston, our machine learning model must predict the house price. To train our machine learning model with boston housing data, we will be using scikit-learn’s boston dataset.

In this dataset, each row describes a boston town or suburb. There are 506 rows and 13 attributes (features) with a target column (price).

```python
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing the Boston Housing dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
# iniciation the dataframe
data = pd.DataFrame(boston.data)
data.head(5)
```

```python
# adding the features names to the dataframe
data.columns = boston.feature_names
data.head(5)
```

Each record in the database describes a Boston suburb or town.

```python
# adding target variable to dataframe
data["PRICE"] = boston.target
# check the shape of the dataframe
data.shape
```

```python
data.info()
```

```python
# Identifying the unique number of values in the dataset
data.nunique()
```

```python
# check for missing values
data.isnull().sum()
```

```python
# finding correlation between the features
corr = data.corr()
corr.shape
```

```python
# plotting heatmap
plt.figure(figsize=(20,20))
sns.heatmap(corr, cbar=True,square=True,fmt=".1f", annot=True, annot_kws={"size":15}, cmap="Greens")
```

```python
corr_price = pd.DataFrame(abs(corr.PRICE).sort_values(ascending=False))
corr_price.style.background_gradient(cmap='viridis')
```

```python
# spliting target variable and independet variables
X = data.drop(["PRICE"],axis=1)
y = data["PRICE"]
```

```python
# splitting to training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y, test_size=0.3,random_state=4)
```

### Linear regresion
training the model

```python
# import library for linear regresion 
from sklearn.linear_model import LinearRegression
#create a linear regresor
lm = LinearRegression()
# train the model using the training test
lm.fit(X_train, y_train)

```

```python
# value of y intecept (donde corta en y la ecuacion lineal b = a  + bx)
lm.intercept_
```

```python
# converting the coefficientes values to a dataframe
coeffcients = pd.DataFrame([X_train.columns,lm.coef_]).T
coeffcients = coeffcients.rename(columns={0:"Atributte",1:"Coeffcients"})
coeffcients.sort_values(by="Coeffcients",ascending=False)
```

### Model evaluation


```python
# predicting model on train data
y_pred = lm.predict(X_train)
```

```python
# model evaluation
print("R^2:", metrics.r2_score(y_train,y_pred))
print("MAE:", metrics.mean_absolute_error(y_train,y_pred))
print("MSE:", metrics.mean_squared_error(y_train,y_pred))
print("RMSE:",np.sqrt( metrics.mean_squared_error(y_train,y_pred)))
```

```python
# visualizing the differences between actual prices and predicted values
plt.scatter(y_train, y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted values")
plt.title("Priced vs Predicted values")
plt.show()
```

```python
# visualizing residuals
plt.scatter(y_pred, y_train-y_pred)
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Predicted vs Residuals")
plt.show()
```

There is no pattern visible in this plot and values are distributed equally around zero. So Linearity assumption is satisfied

```python
# checking normality of errors
#Here the residuals are normally distributed. So normality assumption is satisfied
sns.distplot(y_train-y_pred)
plt.title("Histogram of Residuals")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```
